```python

def calculate_total(prices):
    return sum(prices)

def apply_discount(total):
    if total > 500:
        return total * 0.10
    else:
        return 0

def display_receipt(items, prices, total, discount):
    print("\nReceipt")
    print("-" * 30)
    for name, price in zip(items, prices):
        print(f"{name}: â‚±{price:.2f}")
    print("-" * 30)
    print(f"Total before discount: â‚±{total:.2f}")
    print(f"Discount: â‚±{discount:.2f}")
    print(f"Amount to Pay: â‚±{total - discount:.2f}")
    print("-" * 30)

items = []
prices = []

num_items = int(input("How many items do you want to buy? "))

for i in range(num_items):
    name = input(f"Enter name of item {i+1}: ")
    price = float(input(f"Enter price of item {i+1}: â‚±"))
    items.append(name)
    prices.append(price)

total = calculate_total(prices)
discount = apply_discount(total)
display_receipt(items, prices, total, discount)
```
